## Creating Category IDs

```python
expense_cat = df.loc[df['Transaction Type'] == 'expense']['Category'].unique()
print(len(expense_cat))

expense_id = {}
for x in range(1, len(expense_cat)+1):
    expense_id[expense_cat[x-1]] = x

expense_id
```

```python
income_id = {
    'Salary': 101,
    'Mutual Funds': 102,
    'Stocks': 103
}
income_id
```

```python
ids = {}
ids.update(expense_id)
ids.update(income_id)
ids
```

```python
df = df
df = df.replace({"Category":ids})
df
```
