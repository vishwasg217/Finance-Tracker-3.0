```python
import numpy as np
import pandas as pd
print("Hello World!")
```

```python
df = pd.read_csv('../data/wrangled_data.csv')
df.head()
```

## Creating Category IDs

```python
expense_cat = df.loc[df['Transaction Type'] == 'expense']['Category'].unique()
print(len(expense_cat))

expense_id = {}
for x in range(1, len(expense_cat)+1):
    expense_id[expense_cat[x-1]] = x

expense_id
```

```python
income_id = {
    'Salary': 101,
    'Mutual Funds': 102,
    'Stocks': 103
}
income_id
```

```python
ids = {}
ids.update(expense_id)
ids.update(income_id)
ids
```

```python
df = df
df = df.replace({"Category":ids})
df
```

## Creating Data for the Database

### Category

```python
cat = pd.DataFrame(list(ids.items()),columns = ['category_name','category_id'])

print (cat)
print (type(cat))
```

```python
cat = cat.iloc[:, [1, 0]]
cat
```

```python
cat.insert(2, 'type', np.NaN)

cat['type'][cat['category_id']<=20] = 'expense'
cat['type'][cat['category_id']>20] = 'income'

cat
```

```python
cat.to_csv('../data/database/category.csv', index=False, header=False)
```
